```python
import sqlite3
import json

def _execute(query):
    """Function to execute queries against a local sqlite database"""
    dbPath='data.db'
    connection=sqlite3.connect(dbPath)
    cursorobj=connection.cursor()
    try:
        cursorobj.execute(query)
        result=cursorobj.fetchall()
        connection.commit()
    except Exception:
        raise
    connection.close()
    return result

```

```python
import datetime
import pandas as pd
from pandas.io.sql import read_sql
```

```python
query="select * from rank where currdate='%s'" % datetime.date.today()
_execute(query)
with sqlite3.connect('data.db', detect_types=sqlite3.PARSE_DECLTYPES) as conn:
    df_sqlite = read_sql(query, con=conn)
```

```python
df_sqlite.drop(['currdate','id'],axis=1).T.to_dict().values()
```
